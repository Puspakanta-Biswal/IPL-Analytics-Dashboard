```python
df['toss_win_match_win'] = (df['toss_winner'] == df['winner']).astype(int)
# Create feature: 1 if toss winner also won the match, else 0
```

```python
df.info()   # Check structure
```

```python
df['match_date'] = pd.to_datetime(df['match_date']) # Optimize data types: convert date column to datetime
```

```python
df['team1'] = df['team1'].astype('category') # Optimize data types: categorical columns for efficient processing
df['team2'] = df['team2'].astype('category')
df['winner'] = df['winner'].astype('category')
```

```python
df.info()   # Check structure
```

```python
df.rename(columns={
    'team1': 'team_1',
    'team2': 'team_2',
    'winner': 'match_winner',
    'player_of_match': 'player_of_the_match',
    'team1_players': 'team_1_players',
    'team2_players': 'team_2_players',
    'result': 'match_result',
    'result_margin': 'win_margin'
}, inplace=True)
```

```python
df.head() # View data
```

```python
all_teams = pd.concat([df['team_1'], df['team_2']])
print(all_teams.unique())
```

```python
team_map = {
    'Royal Challengers Bengaluru': 'Royal Challengers Bangalore',
    'Royal Challengers Bangaluru': 'Royal Challengers Bangalore',
    
    'Rising Pune Supergiants': 'Rising Pune Supergiant',
    
    'Delhi Daredevils': 'Delhi Capitals',
    
    'Kings XI Punjab': 'Punjab Kings'
}
```

```python
for col in ['team_1', 'team_2', 'match_winner']:
    df[col] = df[col].replace(team_map)
```

```python
all_teams = pd.concat([df['team_1'], df['team_2']])
print(all_teams.value_counts())
```

```python
X = df[['toss_win_match_win', 'target_runs', 'target_overs']]
y = df['match_winner']
```

```python
from sklearn.preprocessing import LabelEncoder  # Target (convert to numbers)
le = LabelEncoder()

y = le.fit_transform(df['match_winner'])
```

```python
from sklearn.model_selection import train_test_split  # Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression  # Train model
model = LogisticRegression(max_iter=200)

model.fit(X_train, y_train)
```

```python
print("Accuracy:", model.score(X_test, y_test))  # Accuracy
```
